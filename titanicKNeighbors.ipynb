```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier

dta = pd.read_excel('titanic.xls')
dta.info()
```

```python
#dta['age'].isnull().sum()#suma null
dta['age'] = dta['age'].fillna(dta['age'].median())
dta['embarked']=dta['embarked'].fillna('S')
dta.loc[dta['sex'] == 'male', 'sex'] = 0
dta.loc[dta['sex']=='female','sex']=1
dta.loc[dta['embarked']=='S','embarked'] =0
dta.loc[dta['embarked']=='C','embarked'] =1
dta.loc[dta['embarked']=='Q','embarked'] =2
dta['fare'] = dta['fare'].fillna(dta['fare'].median())
```

```python
#new_col = ['age','sex','embarked','pclass','fare','sibsp','parch']
new_col = ['age','sex','embarked','fare','pclass']
x_train = dta[new_col]
y_train = dta['survived']
knn = KNeighborsClassifier()
knn.fit(x_train,y_train)
knn.score(x_train,y_train)
```

```python
a = knn.predict(x_train)
a = a.round()

l=[]
for i in a:
    i = int(i)
    l.append(i)

mainFrame = pd.DataFrame({'passID':dta.index,
                         'survived1': a,
                         'sur':dta['survived']})
mainFrame
```
